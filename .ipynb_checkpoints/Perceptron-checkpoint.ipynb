```python
import numpy as np
import matplotlib.pyplot as plt
```

### Generate Training Data

```python
def arbitrary_function(x, m, b):
    return m*x + b

class Point:
    def __init__(self, m, b):
        #Point takes in an m and b for arbitrary line to label above and below.
        #Initialize points x and y on uniform random distribution between -10 and 10
        self.x = np.random.uniform(-10, 10)
        self.y = np.random.uniform(-10, 10)
        self.bias = 1
        #Points given random x and y if above y=x they are labeled -1 otherwise 1
        self.label = 1 if arbitrary_function(self.x, m, b) < self.y else -1
        
#Initialize and plot 100 points. 
m_test = 9
b_test = -5
num_training_points = 10000


points = [Point(m_test, b_test) for _ in range(num_training_points)]
plt.title('Labeled Training Data')
for point in points:
    col = 'blue' if point.label == 1 else 'red'
    lab = '^' if point.label == 1 else 'o'
    plt.plot(point.x, point.y, color=col, marker=lab)
ys = xs = np.arange(-10, 11)
plt.plot(xs, arbitrary_function(xs,m_test, b_test), 'k')
plt.axis([-10, 10, -10, 10]);
```

### Binary Step Activation Function     

```python
def binary_step(a):
    #Simple binary step activation function. Returns sign of input
    if a >= 0:
        return 1
    else: 
        return -1
```

### Define Perceptron Class

```python
class Perceptron:
    def __init__(self, number_inputs):
        self.learning_rate = 0.1
        #Initialize weights randomly between -1 and 1 for number of inputs.
        self.weights = np.array([np.random.uniform(-1, 1) for _ in range(number_inputs)])
        self.weights = np.append(self.weights, 1) #appends bias
    
    def guess(self, inputs):
        sum = 0
        #multiply inputs by weights and get sign from binary step
        for i,j in enumerate(inputs):
            sum += j*self.weights[i]
        return binary_step(sum)
    
    def train(self, inputs, target):
        guesses = self.guess(inputs)
        #Error is difference between correct value and the guess
        error = target - guesses        
        for i, j in enumerate(self.weights):
            self.weights[i] = j + error*inputs[i]*self.learning_rate
            
        
```

### Initialize a new Perceptron with 2 inputs

```python
x = Perceptron(2)
```

### Train Perceptron on 90% of training data

```python
training_points = points[0:num_training_points - num_training_points//10]
for point in training_points:
    x.train([point.x, point.y, point.bias], point.label)
    p_guess = x.guess([point.x, point.y, point.bias])
    col = 'green' if p_guess == point.label else 'red'
    lab = '^' if point.label == 1 else 'o'
    plt.plot(point.x, point.y, color=col, marker=lab)
plt.plot(xs, arbitrary_function(ys, m_test, b_test), 'k')
plt.axis([-10,10,-10,10]);
```

### Use Remaining 10% of Test Data to Test Trained Perceptron

```python
test_points = points[num_training_points - num_training_points//10:]
for point in test_points:
    p_guess = x.guess([point.x, point.y, point.bias])
    col = 'green' if p_guess == point.label else 'red'
    lab = '^' if point.label == 1 else 'o'
    plt.plot(point.x, point.y, color=col, marker=lab)
plt.plot(xs, arbitrary_function(ys, m_test, b_test), 'k')
plt.axis([-10,10,-10,10]);
```
