```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
```

### Write Logistic Regression Class

```python
class LogisticRegression:
    def __init__(self, xs, ys, epochs = 100, learning_rate=0.01):
        self.learning_rate = learning_rate
        
        self.n = len(xs[0]) + 1
        self.thetas = np.random.uniform(-10, 10, (self.n, 1))
        
        self.xs = np.append(xs, np.ones((np.array(xs).shape[0], 1)), axis=1)
        self.ys = np.array(ys).reshape(len(ys),1)
        
        for _ in range(epochs):
            for x, y in zip(self.xs, self.ys):
                self.train(x, y)
    
    def sigmoid(self, x):
        return 1/(1+np.exp(-x))
    
    def guess (self, val):
        if type(val) in [int, float, np.float64]:
            val = np.array([val,1]).reshape(2,1)
#         print(self.sigmoid(np.matmul(self.thetas.transpose(), val)))
        return self.sigmoid(np.matmul(self.thetas.transpose(), val))
    
    def train(self, x, y):
        self.thetas = self.thetas + self.learning_rate * (y - self.guess(x)) * x.reshape(len(x), 1)
        
    def display(self, step=0.5):
        if len(self.xs[0]) == 2:
            _min = min(list(self.xs[:,0]))
            _max = max(list(self.xs[:,0]))
            plt.scatter(self.xs[:,0], self.ys, color='blue')
            for i in np.arange(_min, _max, step):
                plt.scatter(i, self.guess(i), color='red')
        if len(self.xs[0]) == 3:
            _xs = np.random.uniform(-5, 5, (round(1000*step), 1))
            _ys = np.random.uniform(-5, 5, (round(1000*step), 1))
            X = np.array([np.array([*x,*y]) for x,y in zip(xs, ys)])
            plt.figure()
            plt.title("Training Data Prediction Accuracy")
            for x,y in zip(self.xs, self.ys):
                plt.scatter(x[0], x[1], color='green' if round(*self.guess(x))==y else 'red')
            plt.show()
            plt.figure()
            plt.title("Test Data (purple and orange) Plotted Over (blue and red)")
            for x,y in zip(self.xs, self.ys):
                plt.scatter(x[0], x[1], color = 'red' if y else 'blue')
                
            for x,y in zip(_xs, _ys):
#                 print(x[0],y[0])
                plt.scatter(x, y, color = 'orange' if round(self.guess([*x,*y,1])[0]) else 'purple')
            plt.show()
        
```

### Generate one-dimensional linearly-seperable data set

```python
n = 1000
```

```python
xs = np.random.uniform(-2,2,(n,1))
ys = []
for i in xs:
    ys.append(1 if i > 0.25 else 0)
ys = np.array(ys).reshape(n, 1)
for x,y in zip(xs, ys):
    plt.scatter(x, y, color = 'red' if y else 'blue')
plt.axis([-2,2,-0.05,1.05]);
```

```python
M = LogisticRegression(xs, ys, epochs = 10000)
```

```python
M.display(step=0.01)
```

### Generate 2 dimensional lineraly seperable dataset

```python
xs = np.random.uniform(-5, 5, (n, 1))
ys = np.random.uniform(-5, 5, (n, 1))
labs = []
for x, y in zip(xs, ys):
    labs.append(1 if x-y > 2 else 0)
for x, y, lab in zip(xs, ys, labs):
    plt.scatter(x,y, color = 'red' if lab else 'green')
```

```python
X = np.array([np.array([*x,*y]) for x,y in zip(xs, ys)])
```

```python
L = LogisticRegression(X, labs, epochs = 10000)
```
